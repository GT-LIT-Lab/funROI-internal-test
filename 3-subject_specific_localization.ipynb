The notebook runs the subject specific localization using a given parcels file on a localizer.

Here we use the TripleEvents parcels file instead of the one we generated from the sample data, as in step 2. This parcels file is uploaded into the BIDS parcels directory.

```python
# Import funROI

import sys
import os
package_path = "../funROI"
sys.path.append(os.path.abspath(package_path))
import funROI
from funROI import utils, first_level, analysis

from nilearn.plotting import plot_glass_brain
```

```python
# Set the BIDS data directories
funROI.set_bids_data_folder("./data/bids")
funROI.set_bids_deriv_folder("./data/bids/derivatives")
funROI.set_bids_preprocessed_folder("./data/bids/derivatives/fmriprep")
```

```python
subject_specific_localizer = analysis.SubjectSpecificLocalizer(subject='040', parcels='Events',
                                                               threshold_type='percent', threshold_value=0.1)   # best 10% of voxels in each parcel
roi_img = subject_specific_localizer.localize(localizer='both_sem_minus_perc_or', task='EventsOrig')
```

```python
from nilearn.plotting import plot_glass_brain

plot_glass_brain(roi_img, title='Subject-specific ROI', colorbar=True, display_mode='lyrz')
```
