```python
spcorr_analyzer = analysis.SpatialCorrelationAnalyzer(
    subject = '040',

    # localizer setup if spatial correlation on ssROI is desired 
    localizer_task = 'EventsOrig',
    localizer = 'Sem_sentminusPerc_sent',
    threshold_type = 'percent',
    threshold_value = 0.1,
    parcels = 'Events')

spcorr_analyzer.add_contrast('EventsOrig', 'Sem_sentminusPerc_sent')
spcorr_analyzer.add_contrast('SWNloc', 'S-N')

spcorr_analyzer.run()
```
