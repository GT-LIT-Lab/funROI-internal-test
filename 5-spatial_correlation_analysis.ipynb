The notebook estimates the spatial correlation across different contrasts within subject-specific localized ROIs or parcels.

```python
# Import funROI

import sys
import os
package_path = "../funROI"
sys.path.append(os.path.abspath(package_path))
import funROI
from funROI import utils, first_level, analysis

# Turn off warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Set the BIDS data directories
funROI.set_bids_data_folder("./data/bids")
funROI.set_bids_deriv_folder("./data/bids/derivatives")
funROI.set_bids_preprocessed_folder("./data/bids/derivatives/fmriprep")
```

For subject-specific localized ROIs:

```python
spcorr_analyzer = analysis.SpatialCorrelationAnalyzer(
    subject = '040',
    parcels = 'Events',

    localizer_task = 'EventsOrig',
    localizer = 'Sem_sentminusPerc_sent',
    threshold_type = 'percent',
    threshold_value = 0.1)

spcorr_analyzer.add_contrast('EventsOrig', 'Sem_sentminusPerc_sent')
spcorr_analyzer.add_contrast('SWNloc', 'S-N')

spcorr_analyzer.run()
```

For parcels:

```python
spcorr_analyzer = analysis.SpatialCorrelationAnalyzer(
    subject = '040',
    parcels = 'Events')

spcorr_analyzer.add_contrast('EventsOrig', 'Sem_sentminusPerc_sent')
spcorr_analyzer.add_contrast('SWNloc', 'S-N')

spcorr_analyzer.run()
```
