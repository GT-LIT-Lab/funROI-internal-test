```python
import numpy as np
import pandas as pd
```

```python
spm_ss_results_dir = "../data/spm_ss_results/effects"
funroi_results_dir = "../data/funROI_output/effect"
spm_ss_data = []
funroi_data = []
ii = 0
for localizer in ["language", "events_sent", "events_photo", "events_both"]:
    for effect in ["lang", "events"]:
        dat = pd.read_csv(
            f"{spm_ss_results_dir}/{localizer}_{effect}/spm_ss_mROI_data.csv"
        )
        dat["Localizer"] = localizer
        spm_ss_data.append(dat)
        dat = pd.read_csv(
            f"{funroi_results_dir}/effect_{ii:04d}/effect_summary.csv"
        )
        dat["Localizer"] = localizer
        funroi_data.append(dat)
        ii = ii + 1
spm_ss_data = pd.concat(spm_ss_data)
funroi_data = pd.concat(funroi_data)
sorted_subjects = np.sort(funroi_data["subject"].unique())
spm_ss_data["Subject"] = sorted_subjects[spm_ss_data["Subject"].values - 1]
```

```python
# spm_ss filtered by subject=59
spm_ss_data[(spm_ss_data["Subject"] == 59) & (spm_ss_data["ROI"] == 9)]
```

```python
data = pd.merge(
    spm_ss_data,
    funroi_data,
    left_on=["ROI", "Subject", "Localizer", "Effect"],
    right_on=["froi", "subject", "Localizer", "effect"],
)
data_combined = data[["ROI", "Subject", "Localizer", "Effect"]]
data_combined["spm_ss_effect_size"] = data["EffectSize"]
data_combined["funroi_effect_size"] = data["size"]
data_combined["spm_ss_effect_size - funroi_effect_size"] = (
    data_combined["spm_ss_effect_size"] - data_combined["funroi_effect_size"]
)
# show the data with highest absolute difference
data_combined["abs_diff"] = np.abs(
    data_combined["spm_ss_effect_size - funroi_effect_size"]
)
data_combined.sort_values("abs_diff", ascending=False).head(20)
```
